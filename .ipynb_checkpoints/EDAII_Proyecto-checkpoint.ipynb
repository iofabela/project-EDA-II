```python
#Ejercicio 1
Grafo=[['pantitlan','zaragoza'],['zaragoza','gomezfarias'],['gomezfarias','boulevardpuertoaereo'],['boulevardpuertoaereo','balbuena'],['balbuena','moctezuma'],['montezuma','sanlazaro'],['sanlazaro','candelaria'],['candelaria','merced'],['merced','pinosuarez'],['pinosuarez','isabellacatolica'],['isabellacatolica','saltodelagua'],['saltodelagua','balderas'],['balderas','cuauhtemoc'],['cuauhtemoc','insurgentes'],['insurgentes','sevilla'],['sevilla','chapultepec'],['chapultepec','juanacatlan'],['juanacatlan','tacubaya'],['tacubaya','observatorio'],['cuatrocaminos','panteones'],['panteones','tacuba'],['tacuba','cuitlahuac'],['cuitlahuac','popotla'],['popotla','colegiomilitar'],['colegiomilitar','normal'],['normal','sancosme'],['sancosme','revolucion'],['revolucion','hidalgo'],['hidalgo','bellasartes'],['bellasartes','allende'],['allende','zocalo'],['zocalo','pinosuarez'],['pinosuarez','sanantonioabad'],['sanantonioabad','chabacano'],['chabacano','viaducto'],['viaducto','xola'],['xola','villadecortes'],['villadecortes','nativitas'],['nativitas','portales'],['portales','ermita'],['ermita','generalanaya'],['generalanaya','tasquena'],['indiosverdes','deportivo18demarzo'],['deportivo18demarzo','potrero'],['potrero','laraza'],['laraza','tlatelolco'],['tlatelolco','guerrero'],['guerrero','hidalgo'],['hidalgo','juarez'],['juarez','balderas'],['balderas','ninosheroes'],['ninosheroes','hospitalgeneral'],['hospitalgeneral','centromedico'],['centromedico','etipia/plazadelatransparencia'],['etipia/plazadelatranparencia','eugenia'],['eugenia','diviciondelnorte'],['diviciondelnorte','zapata'],['zapata','coyoacan'],['coyoacan','viveros/derechoshumanos'],['viveros/derechoshumanos','miguelangeldequevedo'],['miguelangeldequevedo','copilco'],['copilco','universidad'],['martincarrera','talisman'],['talisman','bondojito'],['bondojito','consulado'],['consulado','canaldelnorte'],['canaldelnorte','morelos'],['morelos','candelaria'],['candelaria','frayservando'],['frayservando','jamaica'],['jamaica','santaanita'],['pantitlan','hangares'],['hangares','terminalaerea'],['terminalaerea','oceania'],['oceania','aragon'],['aragon','eduardomolina'],['eduardomolina','consulado'],['consulado','vallegomez'],['vallegomez','misterios'],['misterios','laraza'],['laraza','autobusesdelnorte'],['autobusesdelnorte','institutodelpetroleo'],['institutodelpetroleo','politecnico'],['elrosario','tezozomoc'],['tezozomoc','azcapotzalco'],['azcapotzalco','ferreria'],['ferreria','norte45'],['norte45','vallejo'],['vallejo','institutodelpetroleo'],['institutodelpetroleo','lindavista'],['lindavista','deportivo18demarzo'],['deportivo18demarzo','lavilla-basilica'],['lavilla-basilica','martincarrera'],['elrosario','aquilesserdan'],['aquilesserdan','camarones'],['camarones','refineria'],['refineria','tacuba'],['tacuba','sanjoaquin'],['sanjoaquin','polanco'],['polanco','auditorio'],['auditorio','constituyentes'],['constituyentes','tacubaya'],['tacubaya','sanpedrodelospinos'],['sanpedrodelospinos','sanantonio'],['sanantonio','mixcoac'],['mixcoac','barrancadelmuerto'],['garibaldi/lagunilla','bellasartes'],['bellasartes','sanjuanletran'],['sanjuanletran','saltodelagua'],['saltodelagua','doctores'],['doctores','obrera'],['obrera','chabacano'],['chabacano','laviga'],['laviga','santaanita'],['santaanita','coyuya'],['coyuya','iztacalco'],['iztacalco','apatlaco'],['apatlaco','aculco'],['aculco','escuadron201'],['escuadron201','atlalilco'],['atlatilco','iztapalapa'],['iztapalapa','cerrodelaestrella'],['cerrodelaestrella','uami'],['uami','constitucionde1917'],['pantitlan','puebla'],['puebla','ciudaddeportiva'],['ciudaddeportiva','velodromo'],['velodromo','mixiuhca'],['mixiuhca','jamaica'],['jamaica','chabacano'],['chabacano','lazarocardenas'],['lazarocardenas','centromedico'],['centromedico','chilpancingo'],['chilpancingo','patriotismo'],['patritismo','tacubaya'],['pantitlan','agricolaoriental'],['agricolaoriental','canaldesanjuan'],['canaldesanjuan','tepalcates'],['tepalcates','guelatao'],['guelatao','penonviejo'],['penonviejo','acatitla'],['acatitla','santamarta'],['santamarta','losreyes'],['losreyes','lapaz'],['ciudadazteca','plazaaragon'],['plazaaragon','olimpica'],['olimpica','ecatepec'],['ecatepecc','muzquiz'],['muzquiz','riodelosremedio'],['riodelosremedio','impulsora'],['impulsora','nezahualcoyotl'],['nezahualcoyotl','villadearagon'],['villadearagon','bosquedearagon'],['bosquedearagon','deportivooceania'],['deportivooceania','oceania'],['oceania','romerorubio'],['romerorubio','floresmagon'],['floresmagon','sanlazaro'],['sanlazaro','morelos'],['moleros','tepito'],['tepito','garibaldi/lagunilla'],['garibaldi/lagunilla','guerrero'],['guerrero','buenavista'],['tlahuac','tlatenco'],['tlatenco','zapotitlan'],['zapotitlan','nopalera'],['nopalera','alivos'],['olivos','tezonco'],['tezonco','perifericooriente'],['perifericooriente','calle11'],['calle11','tomatlan'],['tomatlan','esimeculhuacan'],['esimeculhuacan','puebloculhuacan'],['puebloculhucan','atlalilco'],['atlalilco','mexicalzingo'],['mexicalzingo','ermita'],['ermita','ejecentral'],['ejecentral','parquedelosvenados'],['parquedelosvenados','zapata'],['zapata','20denoviembre'],['20denoviembre','insurgentessur'],['insurgentesur','mixcoac']]


def Vecino (lista,x):
    sublista=[]    
    for i in range (len(lista)):
        if x==lista[i][0]:
            sublista.append(lista[i][1])
        if x==lista[i][1]:
            sublista.append(lista[i][0])
        sublista.sort()
    return sublista        


print(Vecino(Grafo,'balderas'))


def BPA(G,s):
    subG=[]
    subG.append(s)
    elemento=subG[0]
    G1=Vecino(G,elemento)
    while(G1!=[]):
        G3=[]
        elemento1=G1[0]
        G3=Vecino(G,elemento1)
        tmp=Vecino(G,elemento1)
        
        for i in range(len(G1)):
            for j in range(len(G3)):
                if(G1[i]==G3[j]):
                    elemento2=G1[i]
                    tmp.remove(elemento2)
        for i in range(len(subG)):
            for j in range(len(G3)):
                if(subG[i]==G3[j]):
                    elemento2=subG[i]
                    tmp.remove(elemento2)
        if(tmp!=[]):
            for i in range(len(tmp)):
                G1.append(tmp[i])
        elemento2=G1[0]
        subG.append(elemento2)
        #print("->G1: ",G1)
        G1.pop(0)
        
    return subG




s='balderas'
print("La busqueda de balderas da: ", BPA(Grafo,s))
s='jamaica'
print("La busqueda de jamaica da: ", BPA(Grafo,s))
```
